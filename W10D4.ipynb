#### `Esercizio 3/3`
• Qual è il prezzo medio per i prodotti di colore blu?
• Qual è il prezzo medio per i prodotti di colore rosso o nero?
• Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg?
• Qual è il nome inglese e il costo di produzione (StandardCost) di tutti i
prodotti di taglia 42, peso oltre i 10 Kg e colore argento?
• Visualizziamo lo StandardCost e il DealerPrice degli ultimi 20 elementi
del dataset: quali sono le differenze? C'è un pattern? C'è qualche elemento
che non lo segue?

```python
#prezzo medio per i prodotti di colore blu
filtro = dimproduct["Color"] == "Blue"

prezzo_medio_blue = round(dimproduct.loc[ filtro, "DealerPrice"].mean(), 2)

print(f"Il prezzo medio dei prodotti di colore blue è {prezzo_medio_blue}")
dimproduct.loc[ filtro, "DealerPrice"]
```

```python
#il prezzo medio per i prodotti di colore rosso o nero
filtro_red = dimproduct["Color"] == "Red"
filtro_black = dimproduct["Color"] == "Black"
filtro = filtro_red | filtro_black #|=oppure/or (pipe),  &=and , ~=not

prezzo_medio = round(dimproduct.loc[ filtro, "DealerPrice"].mean(), 2)

print(f"Il prezzo medio dei prodotti di colore rosso o nero è {prezzo_medio}")
dimproduct.loc[ filtro, "DealerPrice"]
```

```python
#il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg

filtro_peso = dimproduct["Weight"] > 10
filtro_taglio = dimproduct["Size"] == "42" #Size è ua stringa contorlo con "dimproduct.dtypes"
filtro = filtro_peso & filtro_taglio

prezzo_max = dimproduct.loc[ filtro, "DealerPrice"].max()

print(f"il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg è di {prezzo_max}")
```

```python
 # il nome inglese e il costo di produzione (StandardCost) di tutti i prodotti di taglia 42, peso oltre i 10 Kg e colore argento

filtro_peso = dimproduct["Weight"] > 10
filtro_taglio = dimproduct["Size"] == "42" 
filtro_colore = dimproduct["Color"] == "Silver"
filtro = filtro_peso & filtro_taglio & filtro_colore

dimproduct.loc[ filtro, ["EnglishProductName", "StandardCost"]]
```

```python
 #lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset: quali sono le differenze? C'è un pattern? C'è qualche elemento che non lo segue?

dimproduct.tail(20).loc[:,["DealerPrice", "StandardCost"]] #: tutta le righe
```

```python
#dati relativi a diverse fasce di età: 20-29, 30-39, 40-50
#la media della pressione sanguigna diastolica

filtro_eta_min = diabetes["Age (years)"] >=20
filtro_eta_max = diabetes["Age (years)"] <=29
filtro = filtro_eta_min & filtro_eta_max

media = round(diabetes.loc[filtro,"Diastolic blood pressure (mm Hg)"].mean(), 2)

print(f"La media della pressione sanguigna diastolica per la fascia di età tra 20 e 29 è {media}")
```

```python

filtro_eta_min = diabetes["Age (years)"] >= 40
filtro_eta_max = diabetes["Age (years)"] <= 49
filtro = filtro_eta_min & filtro_eta_max

media = round(diabetes.loc[filtro,"Diastolic blood pressure (mm Hg)"].mean(), 2)

print(f"La media della pressione sanguigna diastolica per la fascia di età tra 40 e 49 è {media}")
```
