```python
#quante enoteche ci sono, e come si chiamano
lista_enoteche = []
#!termini_da_controllare = ["enoteca", "vino", "calice", "bottiglia"]

for nome in pub["Nome"]:
    if pd.notna(nome): #controlla che un dato non sia nullo (pd.isna (nome) = booleano)
        #!for termine in termini_da_controllare:
        if "enoteca" in nome.lower(): #!if termine in nome.lower():
            lista_enoteche.append(nome)
            print(nome)

print(f"Ci sono, {len(lista_enoteche)}, enoteche")
```

```python
import pandas as pd
```

```python
 insurance = pd.read_csv("/Users/lalainegubaton/Downloads/beginner_datasets/insurance.csv")
insurance
```

```python
print(insurance.shape)
print(insurance.dtypes)
print(insurance.columns)
```

```python
media = round(insurance["charges"].mean(), 2)

print(f"La media di charges è: {media}")

round(insurance.groupby("region")["charges"].mean(), 2)
```

##### Notiamo un'elevata richiesta per l'assicurazione sulla salute nella regione 'Southest', contrario per la regione opposta dove è considerata sotto la media.  Nella zona nord, i costi per le cure mediche sono sulla media, da tener d'occhio la regione 'northwest' dove viene determinata una bassa richiesta sull'assicurazione rispetto alla regione opposta.

```python
media = round(insurance["charges"].mean(), 2)

print(f"La media di charges è: {media}")
print(f"La media di charges rispetto ai sessi:")
round(insurance.groupby("sex")["charges"].mean(), 2)
```

##### Con questi dati, abbiamo determinato la media dei costi per i sessi opposti.

```python
media = round(insurance["charges"].mean(), 2)

print(f"La media di charges è: {media}")
print(f"La media di charges rispoeeto ai smoker:")
round(insurance.groupby("smoker")["charges"].mean(), 2)
```

##### Maggior parte di chi usufruisce l'assicurazione sulla salute sono o erano fumatori.

```python
Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
```

```python
insurance["bmi"].describe() #i descrittori statistici di bmi
```

```python
print(insurance.groupby("bmi")["charges"].agg(["min","max","mean"]).quantile(0.25))
print(insurance.groupby("bmi")["charges"].agg(["min","max","mean"]).quantile(0.50))
print(insurance.groupby("bmi")["charges"].agg(["min","max","mean"]).quantile(0.75))
print(insurance.groupby("bmi")["charges"].agg(["min","max","mean"]).quantile(1))
```

```python
insurance.groupby("bmi")["charges"].describe()
```

```python
insurance.groupby("smoker")["charges"].describe()
```

```python
insurance["bmi_q"] = pd.qcut(insurance["bmi"], 4, labels=["q1", "q2", "q3", "q4"]).astype(str)
insurance.groupby("bmi_q")["charges"].describe()
```
