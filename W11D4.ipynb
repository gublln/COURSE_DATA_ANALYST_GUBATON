#### `Esercizio 2`
Creiamo un DataFrame sintetico, che contiene i guadagni mensili di diverse
annate, con il seguente codice:
years = 5
guadagni = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years),
"Anno": np.repeat(list(range(years)), 12),
"Valore": np.random.randint(800, 5000, 12*years)})
• Calcola la somma cumulativa dei guadagni utilizzando il metodo .cumsum()
• Come sopra, ma raggruppato per ogni anno usando prima un .groupby()

```python
years = 5
guadagni = pd.DataFrame({
    "Mese": list("GFMAMGLASOND"*years), 
    "Anno": np.repeat(list(range(years)), 12), 
    "Valore": np.random.randint(800, 5000, 12*years)
})
```

```python
guadagni["Totale_Cumulativo"] = guadagni["Valore"].cumsum()
guadagni
```

```python
guadagni["Totale_Cumulativo_Annuale"] = guadagni.groupby("Anno")["Valore"].cumsum()
guadagni
```

#### `Esercizio 3 1/2`
Dal database AdventureWorks estraiamo la tabella dimcustomer
• Trasformiamo i nomi dei clienti in modo che abbiano solo lettere minuscole,
e i cognomi in modo che abbiano solo lettere maiuscole
• Sulla colonna EmailAddress, utilizzando il metodo .str.split(),
estraiamo nome utente e dominio
• Sulla colonna Phone, estraiamo ogni parte del numero (ad es. da "1 (11)
500 555-0162" a ["1", "(11)", "500", "555-0162"])
• Utilizzando il metodo .str.contains(), estraiamo tutti gli indirizzi e-mail
che contengono il numero "21"

```python
query = "SELECT * FROM dimcustomer"

dimcustomer = pd.read_sql(query, db_engine)
dimcustomer
```

```python
dimcustomer["FirstName"] = dimcustomer["FirstName"].str.lower()
dimcustomer["LastName"] = dimcustomer["LastName"].str.upper()
dimcustomer
```

```python
#Sulla colonna EmailAddress, utilizzando il metodo .str.split(),estraiamo nome utente e dominio
dimcustomer[["Username", "Domain"]] = dimcustomer["EmailAddress"].str.split("@", expand = True)
dimcustomer[["EmailAddress", "Username", "Domain"]]
```

```python
dimcustomer[["CoutryCode","AreaCode", "Prefix", "LineNumber"]] = dimcustomer["Phone"].str.extract(r"(\d+) (\(\d+\)) (\d+) (\d+-\d+)")
dimcustomer[["Phone", "CoutryCode","AreaCode", "Prefix", "LineNumber"]]
```

```python
dimcustomer.loc[dimcustomer["EmailAddress"].str.contains("21", regex = True)]
```
